```python
# pip install transformers

from transformers import pipeline

# 1. 감정 분석 파이프라인
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)

# 2. 한국어 텍스트 생성 파이프라인
text_generator = pipeline(
    "text-generation",
    model="skt/kogpt2-base-v2",
    tokenizer="skt/kogpt2-base-v2"
)

# 3. 감정 해석 함수
def interpret_sentiment(text):
    result = sentiment_analyzer(text)[0]
    stars = int(result['label'].split()[0])
    score = result['score']

    if stars >= 4:
        sentiment = "😊 긍정"
    elif stars == 3:
        sentiment = "😐 중립"
    else:
        sentiment = "☹️ 부정"

    return sentiment, stars, score

# 4. 감정별 분위기 프롬프트 생성
def build_prompt(text, sentiment):
    if "긍정" in sentiment:
        return f"{text} 햇살이 부드럽게 스며들고, 마음속에는 따뜻한 희망이 피어올랐다."
    elif "중립" in sentiment:
        return f"{text} 그리고 그 일은 조용히 지나갔다. 아무도 크게 신경 쓰지 않았다."
    else:
        return f"{text} 차가운 바람이 스쳐가며, 어둠 속에서 알 수 없는 기척이 다가왔다."

# 5. 감정 기반 스토리 생성
def emotion_based_story(text):
    sentiment, stars, score = interpret_sentiment(text)
    prompt = build_prompt(text, sentiment)
    generated = text_generator(prompt, max_length=80, num_return_sequences=1)[0]['generated_text']

    return {
        "감정 판정": sentiment,
        "별점": stars,
        "신뢰도": round(score, 3),
        "이야기": generated
    }

# 6. 실시간 대화 루프
print("=== 감정별 분위기 스토리텔러 ===")
print("문장을 입력하면 감정 분석 후 분위기에 맞는 이야기를 이어드립니다. '종료' 입력 시 종료.\n")

while True:
    user_input = input("당신의 문장: ")
    if user_input.strip() == "종료":
        print("대화를 종료합니다. 감사합니다!")
        break

    story = emotion_based_story(user_input)
    print(f"\n[감정 분석] {story['감정 판정']} (별점: {story['별점']}, 신뢰도: {story['신뢰도']})")
    print(f"[이야기] {story['이야기']}\n")
```

```python
# pip install transformers

from transformers import pipeline
import random

# 1. 감정 분석 파이프라인
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)

# 2. 텍스트 생성 파이프라인
text_generator = pipeline(
    "text-generation",
    model="skt/kogpt2-base-v2",
    tokenizer="skt/kogpt2-base-v2"
)

# 3. 감정 해석 함수
def interpret_sentiment(text):
    result = sentiment_analyzer(text)[0]
    stars = int(result['label'].split()[0])
    score = result['score']

    if stars >= 4:
        sentiment = "😊 긍정"
    elif stars == 3:
        sentiment = "😐 중립"
    else:
        sentiment = "☹️ 부정"

    return sentiment, stars, score

# 4. 감정별 프롬프트 + 음악 추천
def build_prompt_and_music(text, sentiment):
    if "긍정" in sentiment:
        prompt = f"{text} 햇살이 부드럽게 스며들고, 마음속에는 따뜻한 희망이 피어올랐다."
        music = random.choice([
            "BTS - Magic Shop",
            "IU - Palette",
            "Coldplay - Paradise"
        ])
    elif "중립" in sentiment:
        prompt = f"{text} 그리고 그 일은 조용히 지나갔다. 아무도 크게 신경 쓰지 않았다."
        music = random.choice([
            "Paul Kim - Me After You",
            "Lauv - I Like Me Better",
            "백예린 - Square"
        ])
    else:
        prompt = f"{text} 차가운 바람이 스쳐가며, 어둠 속에서 알 수 없는 기척이 다가왔다."
        music = random.choice([
            "Agust D - The Last",
            "Radiohead - Creep",
            "이하이 - Breathe"
        ])
    return prompt, music

# 5. 감정 기반 스토리 생성
def immersive_story(text):
    sentiment, stars, score = interpret_sentiment(text)
    prompt, music = build_prompt_and_music(text, sentiment)
    generated = text_generator(prompt, max_length=80, num_return_sequences=1)[0]['generated_text']

    return {
        "감정 판정": sentiment,
        "별점": stars,
        "신뢰도": round(score, 3),
        "추천 음악": music,
        "이야기": generated
    }

# 6. 실시간 대화 루프
print("=== 감정 기반 몰입형 스토리텔러 ===")
print("문장을 입력하면 감정 분석 후 분위기 있는 이야기와 음악을 추천합니다. '종료' 입력 시 종료.\n")

while True:
    user_input = input("당신의 문장: ")
    if user_input.strip() == "종료":
        print("대화를 종료합니다. 음악과 이야기로 함께해줘서 고마워요 🎧")
        break

    story = immersive_story(user_input)
    print(f"\n[감정 분석] {story['감정 판정']} (별점: {story['별점']}, 신뢰도: {story['신뢰도']})")
    print(f"[추천 음악] 🎵 {story['추천 음악']}")
    print(f"[이야기] {story['이야기']}\n")
```
